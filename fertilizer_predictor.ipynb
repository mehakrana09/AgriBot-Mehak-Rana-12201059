```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from __future__ import print_function
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn import tree

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./drive/MyDrive/Fertilizer Prediction.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#data processing

from numpy import array
from sklearn.preprocessing import LabelEncoder
```

```python
soil_type = array(df['Soil Type'].tolist())
crop_type = array(df['Crop Type'].tolist())
```

```python
crop_type
```

```python
soil_type
```

```python
label_encoder = LabelEncoder()
integer_encoded1 = label_encoder.fit_transform(soil_type)    #encoding soil type
integer_encoded2 = label_encoder.fit_transform(crop_type)     #encoding crop type
```

```python
integer_encoded1
```

```python
integer_encoded2
```

```python
dict1={}
for i in range(len(integer_encoded1)):
  dict1[soil_type[i]]=integer_encoded1[i]



dict2={}
for i in range(len(integer_encoded2)):
  dict2[crop_type[i]]=integer_encoded2[i]
```

```python
dict1
```

```python
dict2
```

```python
df['Soil Type'] = df['Soil Type'].map(dict1)
df['Crop Type'] = df['Crop Type'].map(dict2)
```

```python
Fertilizer_Name = df['Fertilizer Name']
```

```python
#initilizing LabelEncoder
label_encoder = LabelEncoder()

#fiting label encoder and transform target column
df['Fertilizer Name'] = label_encoder.fit_transform(Fertilizer_Name)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
features = df
target = features.pop('Fertilizer Name')
label = target
```

```python
features
```

```python
label
```

```python
target
```

```python
#empty lists banayenge for making accuracy and model

acc = []
models = []
```

```python
#train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features,target,test_size = 0.2,random_state =2)
```

```python
#decision tree

from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)

DT.fit(X_train,y_train)

predicted_value = DT.predict(X_test)

x = metrics.accuracy_score(y_test, predicted_value)
acc.append(x)
models.append('Decision Tree')
print("DecisionTrees's Accuracy is: ", x*100)

print(classification_report(y_test,predicted_value))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#cross validation score nikalenge from decision tree
score = cross_val_score(DT, features, target, cv=5)
score
```

```python
#gaussian Naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB

NaiveBayes = GaussianNB()

NaiveBayes.fit(X_train,y_train)

predicted_values = NaiveBayes.predict(X_test)
x = metrics.accuracy_score(y_test, predicted_values)
acc.append(x)
models.append('Naive Bayes')
print("Naive Bayes's Accuracy is: ", x)

print(classification_report(y_test,predicted_values))
```

```python
# Cross validation score (NaiveBayes)
score = cross_val_score(NaiveBayes,features,target,cv=5)
score
```

```python
from sklearn.svm import SVC

SVM = SVC(gamma='auto')

SVM.fit(X_train,y_train)

predicted_values = SVM.predict(X_test)

x = metrics.accuracy_score(y_test, predicted_values)
# acc.append(x)
models.append('SVM')
print("SVM's Accuracy is: ", x)

print(classification_report(y_test,predicted_values))
```

```python
# svm classifier

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score

svm_pipeline = make_pipeline(StandardScaler(), SVC(probability=True))
svm_pipeline.fit(X_train, y_train)

# Accuray On Test Data
predictions = svm_pipeline.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test Data: {accuracy*100}%")

acc.append(accuracy)
print()
```

```python
# Cross validation score (SVM)
score = cross_val_score(SVM,features,target,cv=5)
score
```

```python
import pickle
# Dump the trained SVM classifier with Pickle
SVM_pkl_filename = 'SVMClassifier_Fertilizer.pkl'
# Open the file to save as pkl file
SVM_Model_pkl = open(SVM_pkl_filename, 'wb')
pickle.dump(SVM, SVM_Model_pkl)
# Close the pickle instances
SVM_Model_pkl.close()
```
